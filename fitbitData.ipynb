```python
df['Date'] = pd.to_datetime(df['Date']).dt.date

print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Format "Date" column as 'dd-mm-yyyy'
df['Formatted Date'] = df['Date'].dt.strftime('%d-%m-%Y')

df.drop(columns=['Date'], inplace=True)

print(df)
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])

# Custom function to update the year based on the month
def update_year(date):
    if date.month >= 8:
        return date.replace(year=2022)
    else:
        return date

# Apply the custom function to the "Date" column
df['Formatted Date'] = df['Formatted Date'].apply(update_year)

print(df)
```

```python
def change_year_back(row):
    if row['Formatted Date'].month >= 8 and row.name < 10:
        return row['Formatted Date'].replace(year=2023)
    else:
        return row['Formatted Date']

# Apply the custom function to the entire DataFrame
df['Date'] = df.apply(change_year_back, axis=1)

print(df)
```
